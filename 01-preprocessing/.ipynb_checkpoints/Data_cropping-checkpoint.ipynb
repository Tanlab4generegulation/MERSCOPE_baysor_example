```python
import pandas as pd
import numpy as np
```

```python
dat = pd.read_csv('detected_transcripts.csv')
```

```python
dat.describe()
```

```python
dat.head()
```

```python
# match Visualizer coordinates to transcript table
#  dont know better way. scanning each FOV for dimensions
fov_dim = dict()
for f in range(dat['fov'].max()):
    fov = dat.loc[dat['fov'] == f, :]
    dim_x = [int(fov['global_x'].min()), int(fov['global_x'].max())]
    dim_y = [int(fov['global_y'].min()), int(fov['global_y'].max())]
    fov_dim[f] = [dim_x, dim_y]
```

```python
# transcript location:
# trx = (4900.14, 4643.26)
trx = (4819.57, 4567.38)
# search FOVs for matching coordinate space
for f,v in fov_dim.items():
    if int(trx[0]) in range(v[0][0], v[0][1]) and int(trx[1]) in range(v[1][0], v[1][1]):
        hit = f
print(hit)
```

```python
scale = (dat['global_y'][1] - dat['global_y'][0]) / (dat['y'][1] - dat['y'][0])
print(dat.loc[dat['fov']==643, 'global_x'].min()/scale)
print(dat.loc[dat['fov']==686, 'global_x'].max()/scale)
print(dat.loc[dat['fov']==643, 'global_y'].min()/scale)
print(dat.loc[dat['fov']==687, 'global_y'].max()/scale)
```

```python
out = dat[dat['fov'].isin([643, 642, 686, 687])].copy()
```

```python
# convert fov pixels to global pixels
#    642
out.loc[out['fov']==642, ['y']] += 1950
#    643
#      -this is top left (0,0)
#    686
out.loc[out['fov']==686, ['x']] += 1950
#    687
out.loc[out['fov']==687, ['x']] += 1950
out.loc[out['fov']==687, ['y']] += 1950
```

```python
# converting it based on micron conversion
out.loc[:,['global_x']] = out['global_x']/scale
out.loc[:,['global_y']] = out['global_y']/scale
```
