```python
import pandas as pd
import numpy as np
```

```python
dat = pd.read_csv('../data/detected_transcripts.csv')
```

```python
dat.describe()
```

```python
dat.head()
```

```python
# pixel = scale * micron
scale = (dat['y'][1] - dat['y'][0]) / (dat['global_y'][1] - dat['global_y'][0])
scale
```

```python
# convert micron to pixel
dat['global_xpixel'] = dat['global_x'] * scale
dat['global_ypixel'] = dat['global_y'] * scale
```

```python
dat.set_index('Unnamed: 0', inplace=True)
```

```python
dat.to_csv("../data/detected_transcripts_wPixel.csv")
```
