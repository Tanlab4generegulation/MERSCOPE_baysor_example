# Processing and EDA of stitched Baysor segmentation

Just looking at two tiles,

There's decent concordance: ~81% cell ID overlap

Processing Steps:

1. Stitch together the assigned molecules
    - Right cell IDs need to be shifted to be unique
    - Noise will be noise?
    - for overlap, use existing, stitched cell ID
    - *what happens if the merging cell is entirely new?
    - option A
        - If different, greedy assignment on size
    - option B, this is probably faster
        - join segmention dfs on molecule intersect
        - switch stitched molecule if...
            - new molecule assignment confidence is higher, then which cell ID is taken? Compute overlap; split cell or merge... 
            - stiched molecule is noise
            - cell size?
        - the result could be dictionary, that can be used to reassign stitched df values
    - Should I use confidence scores?

### Actually trying it now

```python
stitched_df = dat_list['tile_2'].copy()
for key, segdf in dat_list.items():
    if key == 'tile_2':
        continue
    print('working on... ' + key)
    # adjust the new segmentations cell IDs to be unique from the already stitched results
    seg = segdf.copy()
    seg.loc[:,'cell'] += stitched_df['cell'].max()
    
    # find the overlapping molecules
    mol_overlap = stitched_df.index[np.isin(stitched_df.index, seg.index)]
    if len(mol_overlap) == 0:
        print('no overlap')
        stitched_df = pd.concat([stitched_df, seg])
    else:
        # Find the entire cell associated with those molecules
        df1_cells = np.isin(stitched_df['cell'], stitched_df.loc[mol_overlap, 'cell'])
        df2_cells = np.isin(seg['cell'], seg.loc[mol_overlap, 'cell'])
        df1 = stitched_df.loc[df1_cells, ["cell", "assignment_confidence"]].copy()
        df2 = seg.loc[df2_cells, ["cell", "assignment_confidence"]].copy()
        mol_df = df1.join(df2, how='outer', rsuffix='df2')
        mol_df.fillna(0, inplace=True)

        # all the unique cells 
        stitched_df = pd.concat([stitched_df, seg.loc[~df2_cells, :]])    

        # count how many molecules each of the different cell labels overlap
        cell_overlap = []
        resolve_df = pd.pivot_table(mol_df.loc[:,['cell', 'celldf2']], index='cell', columns='celldf2', aggfunc=len, fill_value=0)
        print(resolve_df.shape)
        r_i = np.argmax(resolve_df.to_numpy(), axis=1) # for each left cell, get majority cell ID from right df 
        # I feel like there should be a better, vector-based approach to this
        for i in tqdm(range(resolve_df.shape[0])):
            l_sum = resolve_df.iloc[i,:].sum() # this is stitched cell size
            r_sum = resolve_df.iloc[:, r_i[i]].sum() # this is joining df cell size
            cell_overlap.append(resolve_df.iloc[i, r_i] / l_sum)
            if l_sum > r_sum:
                # if the joining cell is smaller, should we check for new molecules that could potentially expand the stitch?
                pass
            if r_sum > l_sum:
                # stitch in new molecules, rename to existing molecule cell IDs
                cell1 = resolve_df.index[i]
                cell2 = resolve_df.columns[r_i[i]]
                cell_mol = set(seg.index[seg['cell']==cell2]).difference(stitched_df.index[stitched_df['cell']==cell1])
                # if part of the cell overlap is new, a molecule needs to be switched or added
                stitched_df.loc[cell_mol.intersection(set(stitched_df.index)) , 'cell'] = cell1
                new_mol = cell_mol.difference(set(stitched_df.index))
                if len(new_mol) > 0:
                    add_mol = seg.loc[new_mol, :]
                    add_mol['cell'] = cell1
                    stitched_df = pd.concat([stitched_df, add_mol])
        print('percent overlap:' + str(np.mean(cell_overlap)))
    
```

```python
def pickCellLabel(x, ref, mol_df, thresh = 0.3):
    # if merge is noise...
    if x['celldf2'] == 0:
        return x['cell']
    # if stitched is noise
    if x['cell'] == 0:
        # look up new cell label...or force to noise??
        if ref.loc[:,x['celldf2']].sum() > 75:
            return -1 # I'll deal with it later
        else:
            return x['cell']
    # if contest...
    else:
        # calculate overlap...
        if ref.loc[: , x['celldf2']].sum() / ref.loc[x['cell'] , :].sum() > thresh:
            return x['cell']
        # if poor overlap, ID goes to highest mean confidence
        else:
            cell1_conf = mol_df.loc[mol_df['cell'] == x['cell'], 'assignment_confidence'].mean()
            cell2_conf = mol_df.loc[mol_df['celldf2'] == x['celldf2'], 'assignment_confidencedf2'].mean()
            if cell1_conf > cell2_conf:
                return x['cell']
            else:
                return x['celldf2']
    
```

```python
# Option B
stitched_df = dat_list['tile_2'].copy()
noise_df = stitched_df.loc[stitched_df['cell'] == 0,:]
for key, segdf in dat_list.items():
    if key == 'tile_2':
        continue
    print('working on... ' + key)
    # adjust the new segmentations cell IDs to be unique from the already stitched results -- except for noise!
    seg = segdf.copy()
    # grab the noise
    noise_df_tile = seg.loc[seg['cell'] == 0,].copy()
    print('noise df shape: ', str(noise_df_tile.shape))
    noise_df = pd.concat([noise_df, noise_df_tile])
    seg.loc[seg['cell'] != 0,'cell'] += stitched_df['cell'].max()
    
    # find the overlapping molecules
    mol_overlap = stitched_df.index[np.isin(stitched_df.index, seg.index)]
    if len(mol_overlap) == 0:
        print('no overlap')
        stitched_df = pd.concat([stitched_df, seg])
    else:
        # Find the entire cell associated with those molecules
        
        df1_cells = np.isin(stitched_df['cell'], stitched_df.loc[mol_overlap, 'cell'])
        df2_cells = np.isin(seg['cell'], seg.loc[mol_overlap, 'cell'])
        df1 = stitched_df.loc[df1_cells, ["cell", "assignment_confidence"]].copy()
        df2 = seg.loc[df2_cells, ["cell", "assignment_confidence"]].copy()
        mol_df = df1.join(df2, how='outer', rsuffix='df2')
        mol_df.fillna(0, inplace=True)

        # all the unique cells 
        stitched_df = pd.concat([stitched_df, seg.loc[~df2_cells, :]])    

        resolve_df = pd.pivot_table(mol_df.loc[:,['cell', 'celldf2']], index='cell', columns='celldf2', aggfunc=len, fill_value=0)
        
        # drop the noise confirmed
        
        # mol_df = mol_df.loc[~noise,]
        print('mol df shape: ', str(mol_df.shape))
        print('calculating overlaps for ' + str(mol_df.shape[0]) + ' molecules')
        # figure out which molecules need to be changed and which need to be added
        mol_assign = mol_df.apply(lambda x: pickCellLabel(x, resolve_df, mol_df), axis=1)
        # take care of the hold-overs, goes to larger cell
        hold_mols = np.where(mol_assign == -1)
        if len(hold_mols) > 1:
            cell_counts = mol_assign.value_counts()
            for m in hold_mols:
                if cell_counts[mol_df.iloc[m, 2]] > 75: 
                    mol_assign[m] = mol_df.iloc[m, 2]
                else: 
                    mol_assign[m] = 0
        # overlap estimate
        new_cell_count = len(set(mol_assign).difference(set(mol_df['cell'])))
        print('overlap estimate: ' + str((mol_df.shape[0] - new_cell_count) / mol_df.shape[0]))
        
        # now actually editing the cell label
        replace_mol = set(mol_assign.index).intersection(set(stitched_df.index))
        stitched_df.loc[replace_mol, 'cell'] = mol_assign
        add_mol = set(mol_assign.index).difference(set(stitched_df.index))
        if len(add_mol) > 0:
            new_mol_df = seg.loc[add_mol,:]
            new_mol_df['cell'] = mol_assign.loc[add_mol]
            stitched_df = pd.concat([stitched_df, new_mol_df])
        print('final df shape: ', stitched_df.shape)
        # stitched_df = pd.concat([stitched_df, noise_df])
```
